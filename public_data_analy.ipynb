## 1. tidy data 처리 
- 통계청데이터 / 공공데이터는 모두 요약한 데이터를 제공한다. (위와 같은 형태)
- 이런 형태를 가지고 분석 및 시각화를 하기 위해서는 이 데이터의 형태가 pivot테이블 형태가 아니라 컬럼의 값을 컬럼으로 녹여 tidy data의 형태로 만들어줘야 분석이 가능하다. 

```
import pandas as pd
```

```
# unicode decode error, 한글 인코딩 필요 
df_kosis = pd.read_csv("polulationExample.csv", encoding = "cp949")
df_kosis.shape
```

```
pd.options.display.max_columns = 793 # 모든 컬럼값 확인하고자 한다 
```

```
df_kosis.head() # 분석이 가능한 형태로 변형이 필요 
```

```
df_kosis.head(1)
```

```
df = df_kosis.melt(id_vars = "시군구별") # 시군구별을 id vars로 생각하고 데이터를 녹인다. 
df.head()
```

```
df["시군구별"].unique()
```

```
df[df["시군구별"] == "시군구별"]
```

```
print(df.shape)
df = df[df["시군구별"] != "시군구별"].copy() # 같은 데이터 프레임에 다시 써준다. 
df.shape                                     # 약 200개 정도가 제거됨 
```

```
df["시군구별"].unique()
```

```
df.head()
```

## 2. 전처리
- 시각화 및 분석이 가능한 형태로 분리 (년도와 성별, 월을 분리한다. )

```
df["variable"]
```

```
df["variable"].str.split(".") # [년, 월, 성별 ] 
```

```
df["variable"].str.split(".", expand = True) # 데이터 프레임 형태로 변환 
```

```
df["연도"] = df["variable"].str.split(".", expand = True)[0]
df["월"] = df["variable"].str.split(".", expand = True)[1]
df["성별"] = df["variable"].str.split(".", expand = True)[2] 
df.head()
```

```
df.tail()
```

```
df["성별"].unique()
```

```
df["성별"].nunique()
```

```
df["성별"] = df["성별"].fillna("전체")
df["성별"].unique()
```

```
df["성별"] = df["성별"].replace("1","남자").replace("2","여자")
df["성별"].unique()
```

```
df["성별"].value_counts() # 하나의 변수에 대해 빈도수를 세고자 
```

```
df.head()
```

```
df = df.rename(columns={"variable":"기간", "value":"출생아수"}) # col 이름 변경 
df.head()
```
